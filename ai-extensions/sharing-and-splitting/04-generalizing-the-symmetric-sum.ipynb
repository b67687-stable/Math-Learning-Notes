# Generalizing the Symmetric Sum



> *Stacking, breaking, and rebuilding: or, why the obvious extension fails, why it shouldn't, and the Stirling numbers that bridge the gap*

---



## 1. What We Already Know (A Quick Recap)



The symmetric sum for $k$ is:



$$

1 + 2 + 3 + \cdots + k + \cdots + 3 + 2 + 1 = k^2

$$



We've seen this visually as stacking dots, and we've used it to find the sum of squares.



The trick was: **stack these symmetric sums for $k = 1, 2, \dots, n$, then sum the columns.**



Let's do it carefully for a concrete case, because the exact mechanics matter.



### 1.1 The Concrete Case (n=5)



Stack the symmetric sums:



```

k=1:                         1

k=2:                      1  2  1

k=3:                   1  2  3  2  1

k=4:                1  2  3  4  3  2  1

k=5:             1  2  3  4  5  4  3  2  1

```



Now sum each column. Reading from center out:



| Distance from center | Column sum | Formula |

|---|---|---|

| 0 (center) | $1 + 2 + 3 + 4 + 5$ | $S_1(5) = 15$ |

| 1 | $1 + 2 + 3 + 4$ | $S_1(4) = 10$ |

| 2 | $1 + 2 + 3$ | $S_1(3) = 6$ |

| 3 | $1 + 2$ | $S_1(2) = 3$ |

| 4 | $1$ | $S_1(1) = 1$ |



The total of the whole stack is:



$$

15 + 2 \cdot (10 + 6 + 3 + 1) = 15 + 2 \cdot 20 = 55 = S_2(5)

$$



Which equals $1^2 + 2^2 + 3^2 + 4^2 + 5^2 = 55$. ✓



### 1.2 The General Pattern



Column $j$ (distance $j$ from center) has sum $S_1(n-j)$, where $S_1(m) = 1 + 2 + \dots + m = \frac{m(m+1)}{2}$.



$$

\boxed{S_2(n) = S_1(n) + 2 \sum_{j=1}^{n-1} S_1(n-j)}

$$



This worked because **each symmetric sum of $k$ has center value $k$ and side values $k-1, k-2, \dots, 1$.** When stacked, each column collects values that form an arithmetic sequence — and summing that sequence gave $S_1$.

---



## 2. The Obvious Extension (That Doesn't Quite Work)



If stacking symmetric sums of $k$ gives $S_2(n)$, then stacking symmetric sums of $k^2$ should give $S_3(n)$, right?



Let's try it.



The odd-length symmetric sum of $k^2$ is:



$$

1^2 + 2^2 + 3^2 + \cdots + k^2 + \cdots + 3^2 + 2^2 + 1^2 = \frac{k(2k^2 + 1)}{3}

$$



(This is from the symmetric-sum-of-squares notebook.)



Stack these for $k = 1, 2, \dots, n$:



```

k=1:                                1

k=2:                          1   4   1

k=3:                     1   4   9   4   1

k=4:                1   4   9  16   9   4   1

k=5:           1   4   9  16  25  16   9   4   1

```



Sum the columns (center out):



| Distance from center | Column sum | Formula |

|---|---|---|

| 0 | $1 + 4 + 9 + 16 + 25$ | $S_2(5) = 55$ |

| 1 | $1 + 4 + 9 + 16$ | $S_2(4) = 30$ |

| 2 | $1 + 4 + 9$ | $S_2(3) = 14$ |

| 3 | $1 + 4$ | $S_2(2) = 5$ |

| 4 | $1$ | $S_2(1) = 1$ |



The total is:



$$

S_2(5) + 2 \cdot (S_2(4) + S_2(3) + S_2(2) + S_2(1)) = 55 + 2 \cdot 50 = 155

$$



But we know the total of the stack is ALSO the sum of the symmetric sums themselves:



$$

\sum_{k=1}^5 \frac{k(2k^2+1)}{3} = \frac{2 S_3(5) + S_1(5)}{3}

$$



And $S_3(5) = 1 + 8 + 27 + 64 + 125 = 225$, and $S_1(5) = 15$.

So $\frac{2 \cdot 225 + 15}{3} = \frac{465}{3} = 155$. Yes, the total is 155.



### 2.1 So What's the Problem?



The problem is that **the column sums gave us $S_2$, and the total gave us a formula involving $S_3$.** The column-sum approach doesn't directly give $S_3$ — it gives a relationship:



$$

S_2(n) + 2 \sum_{j=1}^{n-1} S_2(n-j) = \frac{2 S_3(n) + S_1(n)}{3}

$$



If we already know $S_2$ and $S_1$, we can solve for $S_3$:



$$

S_3(n) = \frac{3 \left( S_2(n) + 2\sum_{j=1}^{n-1} S_2(n-j) \right) - S_1(n)}{2}

$$



For $n=5$:

$S_3(5) = (3 \cdot 155 - 15)/2 = (465 - 15)/2 = 225$. ✓



It **does work** — just less directly. Every time we go up one power, we have to solve for it.



### 2.2 Why Isn't It as Clean?



Let's define an operator $T$ that builds a symmetric arrangement:



$$

T(a)(k) = a(k) + 2 \sum_{m=1}^{k-1} a(m)

$$



This is "take sequence $a$, create its symmetric sum centered at $k$."



- $T(k) = k + 2 \sum_{m=1}^{k-1} m = k + k(k-1) = k^2$. ✓

- $T(k^2) = k^2 + 2 \sum_{m=1}^{k-1} m^2 = k^2 + 2 S_2(k-1)$



And $k^2 + 2 S_2(k-1)$ simplifies to $\frac{k(2k^2+1)}{3}$, which is NOT $k^3$.



$$T(k^{p-1}) \neq k^p \quad \text{(except when } p=2 \text{)}$$



The $p=2$ case was a happy coincidence: $T(k) = k^2$ happens to be exactly the next power. This doesn't happen again.



But the stacking method IS giving us something useful — it's giving us a **recursive relationship** between $S_p$, $S_{p-1}$, $S_{p-2}$, .... We just need to solve for $S_p$ each time.



Is there a more direct way? One where the building blocks align naturally?

---



## 3. Breaking Powers Into Their Atomic Parts



Let's try a completely different angle. Instead of wrapping $k^{p-1}$ in a symmetric sum and stacking it, let's **break $k^p$ itself into smaller symmetric pieces** and see what they are.



### 3.1 Starting Small



What's the simplest "symmetric object" we can build from $k$?



- A **point**: $k$ itself. Let's call this $C(k,1) = k$.

- A **triangle** (1D symmetric sum, shifted): $1 + 2 + \cdots + (k-1) = \frac{k(k-1)}{2}$. Let's call this $C(k,2)$.

- A **tetrahedron** (2D symmetric sum, shifted): summing all triangles from 1 to $k-2$: $\displaystyle \sum_{i=1}^{k-2} \frac{i(i+1)}{2} = \frac{k(k-1)(k-2)}{6}$. Let's call this $C(k,3)$.



These are the **binomial coefficients**:



$$

C(k, j) = \frac{k(k-1)(k-2)\cdots(k-j+1)}{j!}

$$



Each $C(k, j)$ is a $(j-1)$-dimensional symmetric object (a simplex).



### 3.2 Now: What Is $k^p$ In Terms of These?



Let's work it out by hand, starting from the top:



$$

\begin{align}

k^1 &= k = C(k,1) \\[4pt]

k^2 &= k + k(k-1) = C(k,1) + 2 \cdot C(k,2) \\[4pt]

k^3 &= k + 3k(k-1) + k(k-1)(k-2)  \\

    &= C(k,1) + 6 \cdot C(k,2) + 6 \cdot C(k,3) \\[4pt]

k^4 &= k + 7k(k-1) + 6k(k-1)(k-2) + k(k-1)(k-2)(k-3) \\

    &= C(k,1) + 14 \cdot C(k,2) + 36 \cdot C(k,3) + 24 \cdot C(k,4)

\end{align}

$$



Wait — how did I get those coefficients? Let me show the work for $k^3$:



We want $k^3 = a \cdot C(k,1) + b \cdot C(k,2) + c \cdot C(k,3)$



$$

\begin{align}

a \cdot k + b \cdot \frac{k(k-1)}{2} + c \cdot \frac{k(k-1)(k-2)}{6} &= k^3 \\

6a k + 3b \, k(k-1) + c \, k(k-1)(k-2) &= 6k^3

\end{align}

$$



Expand and match coefficients for $k$, $k^2$, $k^3$:

- $k^3$: $c = 6$

- $k^2$: $3b - 3c = 0$, so $3b = 18$, $b = 6$

- $k$: $6a - 3b + 2c = 0$, so $6a - 18 + 12 = 0$, $6a = 6$, $a = 1$



So: $k^3 = 1 \cdot C(k,1) + 6 \cdot C(k,2) + 6 \cdot C(k,3)$. ✓



Let's verify with a number, say $k=4$:

$C(4,1)=4$, $C(4,2)=6$, $C(4,3)=4$

$1 \cdot 4 + 6 \cdot 6 + 6 \cdot 4 = 4 + 36 + 24 = 64 = 4^3$. ✓



### 3.3 The Coefficient Pattern



Let's arrange the coefficients in a triangle:



$$

\begin{array}{c|cccccc}

k^p \backslash C(k,j) & j=1 & j=2 & j=3 & j=4 & j=5 & j=6 \\ \hline

k^1 & 1 \\

k^2 & 1 & 2 \\

k^3 & 1 & 6 & 6 \\

k^4 & 1 & 14 & 36 & 24 \\

k^5 & 1 & 30 & 150 & 240 & 120 \\

k^6 & 1 & 62 & 540 & 1560 & 1800 & 720

\end{array}

$$



There's a pattern here. Each number is related to the one above it. Let's look at the ratio of consecutive coefficients in each column, or multiply column $j$ by $j!$:



$$

\begin{array}{c|cccccc}

& j=1 & j=2 & j=3 & j=4 & j=5 & j=6 \\ \hline

k^1: & 1 \\

k^2: & 1 & \mathbf{1} \\

k^3: & 1 & \mathbf{3} & \mathbf{1} \\

k^4: & 1 & \mathbf{7} & \mathbf{6} & \mathbf{1} \\

k^5: & 1 & \mathbf{15} & \mathbf{25} & \mathbf{10} & \mathbf{1} \\

k^6: & 1 & \mathbf{31} & \mathbf{90} & \mathbf{65} & \mathbf{15} & \mathbf{1}

\end{array}

$$



Where each entry is (coefficient) $/ j!$. This is a famous triangle — the **Stirling numbers of the second kind**, $S(p, j)$.



They count how many ways to partition $p$ labeled objects into $j$ non-empty subsets. But here they appear naturally as the coefficients that connect powers to binomial coefficients.

---



## 4. The General Formula



Putting it together:



$$

\boxed{k^p = \sum_{j=1}^{p} S(p,j) \cdot j! \cdot C(k,j)}

$$



Where:

- $S(p,j)$ are the Stirling numbers of the second kind

- $C(k,j) = \frac{k(k-1)\cdots(k-j+1)}{j!}$ are binomial coefficients

- Each $C(k,j)$ is a $(j-1)$-dimensional symmetric object



Now, here's the beautiful part: **summing $C(k,j)$ over $k$ is easy.**



### 4.1 The Hockey-Stick Identity



Binomial coefficients have a telescoping property when summed:



$$

\sum_{k=1}^{n} C(k, j) = C(n+1, j+1)

$$



This is the **hockey-stick identity**. Let's verify for $j=2$, $n=5$:



$\sum_{k=1}^5 C(k,2) = 0 + 1 + 3 + 6 + 10 = 20$

$C(6,3) = 20$. ✓



So every binomial coefficient sum telescopes into a single binomial coefficient!



### 4.2 The Master Formula for Power Sums



Therefore:



$$

\boxed{S_p(n) = \sum_{j=1}^{p} S(p,j) \cdot j! \cdot C(n+1, j+1)}

$$



That's it. Every power sum, expressed directly in terms of binomial coefficients and Stirling numbers.



### 4.3 Verify



**For $p=2$:**



$$

\begin{align}

S_2(n) &= S(2,1) \cdot 1! \cdot C(n+1,2) + S(2,2) \cdot 2! \cdot C(n+1,3) \\

       &= 1 \cdot 1 \cdot \frac{n(n+1)}{2} + 1 \cdot 2 \cdot \frac{n(n+1)(n-1)}{6} \\

       &= \frac{n(n+1)}{2} + \frac{n(n+1)(n-1)}{3} \\

       &= \frac{n(n+1)(2n+1)}{6}

\end{align}

$$



**For $p=3$:**



$$

\begin{align}

S_3(n) &= 1 \cdot 1 \cdot C(n+1,2) + 3 \cdot 2 \cdot C(n+1,3) + 1 \cdot 6 \cdot C(n+1,4) \\

       &= C(n+1,2) + 6 \cdot C(n+1,3) + 6 \cdot C(n+1,4)

\end{align}

$$



For $n=5$:

$C(6,2)=15$, $C(6,3)=20$, $C(6,4)=15$

$S_3(5) = 15 + 6 \cdot 20 + 6 \cdot 15 = 15 + 120 + 90 = 225$. ✓



**For $p=4$:**



$$

S_4(n) = C(n+1,2) + 14 \cdot C(n+1,3) + 36 \cdot C(n+1,4) + 24 \cdot C(n+1,5)

$$



For $n=3$: $C(4,2)=6$, $C(4,3)=4$, $C(4,4)=1$, $C(4,5)=0$

$S_4(3) = 6 + 14 \cdot 4 + 36 \cdot 1 = 6 + 56 + 36 = 98$

$1^4 + 2^4 + 3^4 = 1 + 16 + 81 = 98$. ✓

## 5. What Does This Have To Do With Symmetric Sums?



Everything. Look at what each $C(k,j)$ really is:



- $C(k,1) = k$ — a single point (0 dimensions)

- $C(k,2) = \frac{k(k-1)}{2} = 1 + 2 + \cdots + (k-1)$ — a 1-dimensional triangle

- $C(k,3) = \frac{k(k-1)(k-2)}{6} = \sum_{i=1}^{k-2} \sum_{j=1}^{i} j$ — a 2-dimensional tetrahedron



Each $C(k,j)$ IS a symmetric sum — just in higher dimensions!



And when you stack these and sum columns, you get $C(k, j-1)$ — the same telescoping pattern you saw with $S_2$.



### 5.1 The "Stack and Sum Columns" Method, General Version



For any $j$, stack $C(k,j)$ for $k = 1, 2, \dots, n$:



```

C(1, j)

C(2, j)

C(3, j)

...

C(n, j)

```



The $j$-dimensional symmetric arrangement of $C(k,j)$ has a structure where each "column" sum is $C(n-m+1, j-1)$ — exactly like how the 1D symmetric sum of $k$ had column sums $S_1(n-j)$.



The total is $S_1$ for the 1D case, $S_2$ for the 2D case, and in general:



$$

\sum_{k=1}^n C(k,j) = C(n+1, j+1)

$$



### 5.2 The Two Pathways



We now have **two ways** to reach any $S_p(n)$:



**Pathway A — The Recursive Stack** (what we tried in Section 2):

- Stack the symmetric sums of $k^{p-1}$

- Column sums give $S_{p-1}(n-j)$

- Total relates $S_p$ to $S_{p-1}$ and lower

- Solve for $S_p$ — works, but requires knowing lower sums first



**Pathway B — The Stirling Decomposition** (what we found in Section 3-4):

- Decompose $k^p$ into $C(k,1)$ through $C(k,p)$

- Each $C(k,j)$ telescopes to $C(n+1, j+1)$ when summed

- Combine with Stirling coefficients → direct formula for $S_p(n)$



Pathway A is how you'd naturally extend your `sum-of-squares-with-stack-of-symmetric-sums` notebook.

Pathway B is the explicit formula that Pathway A converges to.



### 5.3 A Numeric Comparison



Let's see both paths for $S_4(5)$:



**Pathway A:** Stack symmetric sums of $k^3$ (each equals $\frac{k^2(k^2+1)}{2}$):

- Column sums: $S_3(5-j)$

- Total $= S_3(5) + 2\sum_{j=1}^4 S_3(5-j) = 225 + 2(100+36+9+1) = 225 + 292 = 517$

- This also equals $\sum \frac{k^2(k^2+1)}{2} = \frac{S_4(5) + S_2(5)}{2}$

- So $S_4(5) = 2 \cdot 517 - 55 = 1034 - 55 = 979$. ✓



**Pathway B:**

$S_4(5) = C(6,2) + 14\cdot C(6,3) + 36\cdot C(6,4) + 24\cdot C(6,5)$

$= 15 + 14\cdot20 + 36\cdot15 + 24\cdot6 = 15 + 280 + 540 + 144 = 979$.



Both pathways give $979 = 1^4 + 2^4 + 3^4 + 4^4 + 5^4$. ✓

## 6. Python Playground



Let's build the Stirling triangle and the power sum formula from scratch.

````python
def stirling2(p, j):

    """Stirling numbers of the second kind: S(p,j)"""

    if p == j:

        return 1

    if j == 0 or p == 0:

        return 0

    return j * stirling2(p-1, j) + stirling2(p-1, j-1)



# Display the triangle

print("Stirling numbers of the second kind S(p,j):")

print("p\\j", end="")

for j in range(1, 8):

    print(f"{j:6d}", end="")

print()

for p in range(1, 9):

    print(f"{p}  ", end="")

    for j in range(1, p+1):

        print(f"{stirling2(p, j):6d}", end="")

    print()
````

````python
import math



def binom(n, k):

    """Binomial coefficient C(n,k)"""

    if k < 0 or k > n:

        return 0

    return math.comb(n, k)



def S(p, n):

    """Sum of k^p for k=1..n, using the Stirling formula"""

    total = 0

    for j in range(1, p+1):

        coeff = stirling2(p, j) * math.factorial(j)

        total += coeff * binom(n+1, j+1)

    return total



# Verify against direct computation

print("Verifying the formula for various p, n:")

for p in range(1, 7):

    for n in range(1, 10):

        direct = sum(k**p for k in range(1, n+1))

        formula = S(p, n)

        assert direct == formula, f"MISMATCH p={p}, n={n}: {direct} vs {formula}"

    print(f"  p={p}: all n=1..9 verified ✓")



print()

print("\nNow let's expand the formula for each power:")

for p in range(1, 7):

    terms = []

    for j in range(1, p+1):

        coeff = stirling2(p, j) * math.factorial(j)

        if coeff == 1:

            terms.append(f"C(n+1,{j+1})")

        else:

            terms.append(f"{coeff}·C(n+1,{j+1})")

    print(f"  S_{{{p}}}(n) = {' + '.join(terms)}")
````

````python
def S_recursive(p, n, memo={}):

    """Compute S_p(n) recursively using the stacking method (Pathway A).

    This is the "solve for S_p" approach from Section 2."""

    if p == 1:

        return n * (n + 1) // 2

    if (p, n) in memo:

        return memo[(p, n)]

    

    # First compute all lower S's

    S_prev = [S_recursive(p-1, m) for m in range(1, n+1)]

    

    # Stack and column-sum approach:

    # The total of the stack = S_{p-1}(n) + 2·Σ S_{p-1}(n-j)

    stack_total = S_prev[n-1] + 2 * sum(S_prev[n-j-1] for j in range(1, n))

    

    # From the symmetric sum formula, the odd-length symmetric sum of k^{p-1}

    # can be expressed in terms of S_p, S_{p-2}, ... We need the specific formula.

    # For demonstration, we use the direct formula instead (from the notebook).

    memo[(p, n)] = S(p, n)  # fallback to the proven formula

    return memo[(p, n)]



# Actually, let's just show both pathways produce the same results:

print("Comparing both pathways for S_3(n):")

print("n  | Stack total (A) | S_3 via solve (A) | Stirling formula (B) | Direct")

print("-" * 65)

for n in range(1, 11):

    # Pathway A: stack symmetric sums of k^2

    S2_vals = [S(2, m) for m in range(1, n+1)]

    stack_A = S2_vals[n-1] + 2 * sum(S2_vals[n-j-1] for j in range(1, n))

    # Solve: stack_A = (2·S_3(n) + S_1(n))/3

    S1_n = S(1, n)

    S3_A = (3 * stack_A - S1_n) // 2

    

    # Pathway B: Stirling formula

    S3_B = S(3, n)

    

    # Direct

    direct = sum(k**3 for k in range(1, n+1))

    

    print(f"{n:2d} | {stack_A:12d} | {S3_A:18d} | {S3_B:18d} | {direct}")

    assert S3_A == S3_B == direct

print()

print("Both pathways agree! ✓")
````

````python
# Let's explore: compute S_p(n) for a range of p,n

print("S_p(n) table (p rows, n columns):")

print("p\\n", end="")

for n in range(1, 11):

    print(f"{n:8d}", end="")

print()

for p in range(1, 9):

    print(f"{p}  ", end="")

    for n in range(1, 11):

        print(f"{S(p, n):8d}", end="")

    print()
````

## 7. What We Learned



1. **The stacking method (Section 2) DOES generalize** — it just becomes recursive. Stacking symmetric sums of $k^{p-1}$ gives a relationship that lets you solve for $S_p$, but you need $S_{p-1}, S_{p-2}, \dots$ first.



2. **The clean generalization is the Stirling decomposition (Section 3-4).** Every power $k^p$ breaks into binomial coefficients $C(k,j)$, which ARE $(j-1)$-dimensional symmetric objects. Summing them telescopes via the hockey-stick identity, giving an explicit formula for $S_p(n)$.



3. **The two paths are the same.** The recursive stacking method, unrolled completely, produces exactly the Stirling/binomial formula. They're not competing explanations — they're the same mechanism seen from different altitudes.



4. **The $p=2$ case was a trap.** Because $k^2$ happens to be both a 1D symmetric sum AND a power, it made the generalization look simpler than it actually is. The real structure only appears when you look at $p=3$ and above.



> The symmetric sum method isn't "wrong" for higher powers — it just needs higher dimensions. And those higher-dimensional symmetric sums are exactly the binomial coefficients $C(k,j)$.
