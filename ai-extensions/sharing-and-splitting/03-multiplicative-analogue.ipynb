# The Multiplicative Analogue: Uniform Factor Sharing

> *The additive sharing patterns used arithmetic sequences. The multiplicative analogue uses geometric sequences — and requires perfect powers.*

---

## 1. The Core Idea

If addition has partition regrouping, multiplication has **factor regrouping**:

$$
a \cdot b \cdot c \cdot \cdots \cdot g = (b \cdot a_1) \cdot (c \cdot a_2) \cdot \cdots \cdot (g \cdot a_{n-1})
$$

where $a = a_1 \cdot a_2 \cdot \cdots \cdot a_{n-1}$. The product is preserved, the factor count drops from $n$ to $n-1$.

But unlike addition, **any factorization works** — because multiplication is commutative and associative, $a \cdot b \cdot c$ can be rewritten as $(b \cdot a_1) \cdot (c \cdot a_2)$ for literally any $a_1, a_2$ whose product is $a$. There's no constraint, no interesting structure, nothing to discover.

**The interesting case is uniform factor sharing**: splitting $a$ into $n-1$ *equal* factors.

---

## 2. Uniform Factor Sharing

Split $a$ into $n-1$ equal factors: $a = q^{n-1}$, where $q = a^{1/(n-1)}$.

For a geometric progression $a, ar, ar^2, \dots, ar^{n-1}$:

$$
a \cdot ar \cdot ar^2 \cdot \cdots \cdot ar^{n-1}
    = (ar \cdot q) \cdot (ar^2 \cdot q) \cdot \cdots \cdot (ar^{n-1} \cdot q)
$$

The result is also a geometric progression with the same ratio $r$, shifted shorter by one term. For integer results, $a$ must be a perfect $(n-1)$-th power.

### Row Structure

Let $m$ be the row number, with $m+1$ factors on the left and $m$ on the right ($n = m+1$).
The first term must be an $m$-th power: $a = q^m$.

$$
\boxed{
q^m \cdot (q^m r) \cdot (q^m r^2) \cdot \cdots \cdot (q^m r^m)
    = (q^{m+1} r) \cdot (q^{m+1} r^2) \cdot \cdots \cdot (q^{m+1} r^m)
}
$$

Left: $m+1$ factors, first $= q^m$, ratio $r$

Right: $m$ factors, first $= q^{m+1} r$, ratio $r$

Verification (product of geometric sequence):

$$
\begin{align}
\text{LHS} &= q^{m(m+1)} \cdot r^{0+1+\cdots+m} = q^{m(m+1)} \cdot r^{m(m+1)/2} \\
\text{RHS} &= q^{m(m+1)} \cdot r^{1+2+\cdots+m} = q^{m(m+1)} \cdot r^{m(m+1)/2}
\end{align}
$$

Both sides equal. ✓

---

## 3. Pattern Families

### Family A: $q = r$ (the simplest)

When the base $q$ equals the ratio $r$, every term is a pure power of $q$.

**Example: $q=2, r=2$**

$$
\begin{align}
    2 \cdot 4 &= 8 \\
    4 \cdot 8 \cdot 16 &= 16 \cdot 32 \\
    8 \cdot 16 \cdot 32 \cdot 64 &= 32 \cdot 64 \cdot 128 \\
    &\vdots
\end{align}
$$

Row $m$: $2^m$ to $2^{2m}$ on left, $2^{2m+1}$ to $2^{3m}$ on right.

**Example: $q=3, r=3$**

$$
\begin{align}
    3 \cdot 9 &= 27 \\
    9 \cdot 27 \cdot 81 &= 81 \cdot 243 \\
    27 \cdot 81 \cdot 243 \cdot 729 &= 243 \cdot 729 \cdot 2187 \\
    &\vdots
\end{align}
$$

### Family B: $q \neq r$

**Example: $q=2, r=3$**

$$
\begin{align}
    2 \cdot 6 &= 12 \\
    4 \cdot 12 \cdot 36 &= 24 \cdot 72 \\
    8 \cdot 24 \cdot 72 \cdot 216 &= 48 \cdot 144 \cdot 432 \\
    &\vdots
\end{align}
$$

**Example: $q=3, r=2$**

$$
\begin{align}
    3 \cdot 6 &= 18 \\
    9 \cdot 18 \cdot 36 &= 54 \cdot 108 \\
    27 \cdot 54 \cdot 108 \cdot 216 &= 162 \cdot 324 \cdot 648 \\
    &\vdots
\end{align}
$$

---

## 4. Consecutive Integers (Perfect Powers Needed)

For consecutive integers, uniform factor sharing requires the first term to be a perfect power.

### 3 factors → 2 factors (first term must be a perfect square)

$$
\begin{align}
    4 \cdot 5 \cdot 6 &= 10 \cdot 12 \\
    9 \cdot 10 \cdot 11 &= 30 \cdot 33 \\
    16 \cdot 17 \cdot 18 &= 68 \cdot 72 \\
    25 \cdot 26 \cdot 27 &= 130 \cdot 135 \\
    &\vdots
\end{align}
$$

Each row: $n^2 \cdot (n^2+1) \cdot (n^2+2) = (n(n^2+1)) \cdot (n(n^2+2))$

### 4 factors → 3 factors (first term must be a perfect cube)

$$
\begin{align}
    8 \cdot 9 \cdot 10 \cdot 11 &= 18 \cdot 20 \cdot 22 \\
    27 \cdot 28 \cdot 29 \cdot 30 &= 84 \cdot 87 \cdot 90 \\
    64 \cdot 65 \cdot 66 \cdot 67 &= 260 \cdot 264 \cdot 268 \\
    &\vdots
\end{align}
$$

Each row: $n^3 \cdot (n^3+1) \cdot (n^3+2) \cdot (n^3+3) = (n(n^3+1)) \cdot (n(n^3+2)) \cdot (n(n^3+3))$

This connects perfect powers to consecutive integers — the same kind of number-theoretic bridge that the additive patterns made with squares, hexagonals, and primes.

---

## 5. Python Verification

```python
import math

def uniform_geometric(q, r, m):
    """Generate the uniform factor sharing identity for geometric sequence."""
    left = [q**m * (r**i) for i in range(m+1)]
    right = [q**(m+1) * (r**i) for i in range(1, m+1)]
    return left, right

def show_geometric_family(name, q, r, rows=5):
    print(f"\n=== {name} (q={q}, r={r}) ===")
    for m in range(1, rows+1):
        left, right = uniform_geometric(q, r, m)
        lhs = ' \u00b7 '.join(map(str, left))
        rhs = ' \u00b7 '.join(map(str, right))
        ok = math.prod(left) == math.prod(right)
        print(f"  Row {m}: {lhs} = {rhs}  {'\u2713' if ok else '\u2717'}")

def uniform_consecutive(n, k):
    """Uniform factor sharing for k consecutive integers starting at n^k"""
    return [n**k + i for i in range(k+1)]

print("Each identity is verified automatically.")

show_geometric_family("Pure powers", q=2, r=2)
show_geometric_family("Mixed ratio", q=2, r=3)
show_geometric_family("Mixed ratio", q=3, r=2)

print("\n=== Consecutive Integers ===")
print("3 terms (perfect squares):")
for n in range(2, 7):
    terms = uniform_consecutive(n, 2)
    q = n
    right = [(n**2 + 1) * q, (n**2 + 2) * q]
    lhs = ' \u00b7 '.join(map(str, terms))
    rhs = ' \u00b7 '.join(map(str, right))
    ok = math.prod(terms) == math.prod(right)
    print(f"  {lhs} = {rhs}  {'\u2713' if ok else '\u2717'}")

print("\n4 terms (perfect cubes):")
for n in range(2, 5):
    terms = uniform_consecutive(n, 3)
    q = n
    right = [(n**3 + 1) * q, (n**3 + 2) * q, (n**3 + 3) * q]
    lhs = ' \u00b7 '.join(map(str, terms))
    rhs = ' \u00b7 '.join(map(str, right))
    ok = math.prod(terms) == math.prod(right)
    print(f"  {lhs} = {rhs}  {'\u2713' if ok else '\u2717'}")
```

---

## Key Takeaways

1. **Arbitrary factor sharing is trivial** — any split works by commutativity. Nothing to see there.

2. **Uniform factor sharing is the real analogue** — splitting the first term into equal factors requires perfect powers, creating the same kind of structure as the additive patterns.

3. **Geometric sequences replace arithmetic sequences** — in addition, the nice patterns come from arithmetic progressions with common difference $d$. In multiplication, they come from geometric progressions with common ratio $r$.

4. **The parallel is exact**:
   - Additive: $a = kd(n-1)$, result preserves arithmetic progression
   - Multiplicative: $a = q^{n-1}$, result preserves geometric progression

5. **Perfect powers connect to consecutive integers** — uniform factor sharing with consecutive integers reproduces the number-theoretic character of the additive patterns.
