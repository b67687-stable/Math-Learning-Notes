# The Multiplicative Analogue: Factor Sharing

> *What happens when the same "split and share" idea is applied to multiplication instead of addition?*

---

## 1. The Core Idea

If addition has "partition regrouping" (split one term into additive parts, distribute), then multiplication has **factor regrouping**:

- Take a product of $n$ factors
- Split one factor into $n-1$ **factor parts**
- Multiply each part into one of the other $n-1$ factors
- The product is preserved, but the factor count drops from $n$ to $n-1$

Formally:

$$
\underbrace{a \times b \times c \times \cdots \times g}_{\text{n factors}}
    = \underbrace{(b \times a_1) \times (c \times a_2) \times \cdots \times (g \times a_{n-1})}_{\text{n-1 factors}}
$$

where $a = a_1 \times a_2 \times \cdots \times a_{n-1}$.

Since multiplication is commutative, the order of distribution doesn't matter.

---

## 2. Key Difference from Addition

In addition, splitting $a$ into $n-1$ parts is unique only if we require equal parts. In multiplication, splitting $a$ into $n-1$ factors is **never unique** — there are infinitely many factorizations of any integer!

This means the multiplicative analogue is **richer**: each input product can be shared in multiple distinct ways, producing completely different-looking identities.

---

## 3. Example: Consecutive Integers (3 factors → 2 factors)

### Example A: $2 \times 3 \times 4$

$$
2 \times 3 \times 4 = \begin{cases}
    3 \times 8 & (2 = 2 \times 1) \\
    6 \times 4 & (2 = 1 \times 2)
\end{cases}
$$

Both equal $24$. ✓

### Example B: $4 \times 5 \times 6$

$$
4 \times 5 \times 6 = \begin{cases}
    10 \times 12 & (4 = 2 \times 2) \\
    5 \times 24 & (4 = 1 \times 4) \\
    20 \times 6 & (4 = 4 \times 1)
\end{cases}
$$

All equal $120$. ✓

### Example C: $6 \times 7 \times 8$

$$
6 \times 7 \times 8 = \begin{cases}
    7 \times 48 & (6 = 6 \times 1) \\
    14 \times 24 & (6 = 2 \times 3) \\
    21 \times 16 & (6 = 3 \times 2) \\
    42 \times 8 & (6 = 1 \times 6) \\
    12 \times 28 & (6 = 3 \times 2,\ \text{other order})
\end{cases}
$$

All equal $336$. ✓

## 4. A "Natural Numbers" Multiplicative Pattern

Let's fix the first factor split as the **trivial factorization** $a = a \times 1$ (so $a_1 = a$, $a_2 = 1$, etc.):

$$
n \times (n+1) \times (n+2) = (n+1) \times (n+2)n
$$

For consecutive triples:

$$
\begin{align}
    2 \times 3 \times 4 &= 3 \times 8 \\
    3 \times 4 \times 5 &= 4 \times 15 \\
    4 \times 5 \times 6 &= 5 \times 24 \\
    5 \times 6 \times 7 &= 6 \times 35 \\
    6 \times 7 \times 8 &= 7 \times 48 \\
    &\vdots
\end{align}
$$

Right side: $(n+1) \times n(n+2)$ — always equal to the original product.

If instead we split evenly ($a = \sqrt{a} \times \sqrt{a}$):

$$
\begin{align}
    4 \times 5 \times 6 &= 10 \times 12 \\
    9 \times 10 \times 11 &= 30 \times 33 \\
    16 \times 17 \times 18 &= 68 \times 72 \\
    &\vdots
\end{align}
$$

(This requires the first factor to be a perfect square.)

---

## 5. Geometric Sequence Analogue

For a geometric progression $a, ar, ar^2, \dots, ar^{n-1}$:

Split $a$ into $n-1$ equal factors: $q = a^{1/(n-1)}$.

$$
a \times ar \times ar^2 \times \cdots \times ar^{n-1}
    = (ar \cdot q) \times (ar^2 \cdot q) \times \cdots \times (ar^{n-1} \cdot q)
$$

**Proof:**

$$
\begin{align}
\text{LHS} &= a^n \cdot r^{0+1+2+\cdots+(n-1)} = a^n \cdot r^{n(n-1)/2} \\
\text{RHS} &= (ar \cdot q)(ar^2 \cdot q)\cdots(ar^{n-1} \cdot q) \\
           &= a^{n-1} \cdot r^{n(n-1)/2} \cdot q^{n-1} \\
           &= a^{n-1} \cdot r^{n(n-1)/2} \cdot a = a^n \cdot r^{n(n-1)/2} = \text{LHS}
\end{align}
$$

For integer $q$, $a$ must be a perfect $(n-1)$-th power.

### Example: $4 \times 8 \times 16$ (3 terms, $a=4$, $r=2$, $n=3$)

$$
4 \times 8 \times 16 = (8 \times 2) \times (16 \times 2) = 16 \times 32 = 512
$$

All terms are powers of 2. ✓

### Example: $9 \times 27 \times 81$ (3 terms, $a=9$, $r=3$, $n=3$)

$$
9 \times 27 \times 81 = (27 \times 3) \times (81 \times 3) = 81 \times 243 = 19683
$$

## 6. Python Verification

Let's explore the multiplicative analogue computationally.

```python
import math

def multiplicative_share_triple(a, b, c, split_mode="trivial"):
    """Split 'a' into 2 factors and share to b and c."""
    if split_mode == "trivial":
        a1, a2 = a, 1
    elif split_mode == "even":
        sqrt_a = int(math.isqrt(a))
        if sqrt_a * sqrt_a != a:
            return None, None, f"{a} is not a perfect square"
        a1, a2 = sqrt_a, sqrt_a
    elif split_mode == "factor_pair":
        # Use the largest divisor <= sqrt(a)
        for d in range(int(math.isqrt(a)), 0, -1):
            if a % d == 0:
                a1, a2 = d, a // d
                break
    
    new_b, new_c = b * a1, c * a2
    original = a * b * c
    result = new_b * new_c
    
    return (new_b, new_c), original, result

print("=== Consecutive triples (trivial split a = a × 1) ===")
for n in range(2, 9):
    a, b, c = n, n+1, n+2
    (nb, nc), orig, res = multiplicative_share_triple(a, b, c, "trivial")
    print(f"  {a}×{b}×{c} = {nb}×{nc}  (orig={orig}, result={res}) {'✓' if orig==res else '✗'}")

print("\n=== Consecutive triples where first factor is a perfect square (even split) ===")
for a in [4, 9, 16, 25, 36]:
    b, c = a+1, a+2
    (nb, nc), orig, res = multiplicative_share_triple(a, b, c, "even")
    if nb is not None:
        print(f"  {a}×{b}×{c} = {nb}×{nc}  (orig={orig}, result={res}) {'✓' if orig==res else '✗'}")
    else:
        print(f"  {a}×{b}×{c}: skipped ({res})")
```

```python
def geometric_factor_share(a, r, n, split_mode="even"):
    """
    Geometric series: a, ar, ar², ..., arⁿ⁻¹ (n terms).
    Split a into n-1 equal factors.
    """
    from decimal import Decimal, getcontext
    getcontext().prec = 50
    
    # Check if a is a perfect (n-1)-th power
    q = round(a ** (1/(n-1)))
    if q ** (n-1) != a:
        return None, None, f"{a} is not a perfect {n-1}-th power"
    
    terms = [a * (r ** i) for i in range(n)]
    new_terms = [terms[i] * q for i in range(1, n)]
    
    original_product = 1
    for t in terms:
        original_product *= t
    
    new_product = 1
    for t in new_terms:
        new_product *= t
    
    return new_terms, original_product, new_product

print("=== Geometric Factor Sharing ===")
for (a, r, n) in [(4, 2, 3), (9, 3, 3), (8, 2, 4), (16, 2, 3)]:
    new_terms, orig, res = geometric_factor_share(a, r, n)
    if new_terms is not None:
        terms_str = " × ".join(map(str, new_terms))
        print(f"  a={a}, r={r}, n={n}: → {terms_str}  ({orig} = {res}) {'✓' if orig==res else '✗'}")
    else:
        print(f"  a={a}, r={r}, n={n}: skipped ({res})")
```

---

## Key Takeaways

1. **The multiplicative analogue exists and is valid**: factor splitting preserves products just as partition splitting preserves sums.
2. **It's richer than addition**: because factorization is not unique, the same input can be shared in multiple distinct ways.
3. **The consecutive integer case** produces a simple family: $n \times (n+1) \times (n+2) = (n+1) \times n(n+2)$.
4. **The geometric case** mirrors the additive arithmetic case, but requires perfect powers for integer results.
5. **The mechanism is the same**: split something into pieces, give each piece to a surviving term, the total is preserved but the count drops by one.

## Bonus: Going From 4 Factors to 3

$$
a \times b \times c \times d \rightarrow (b \times a_1) \times (c \times a_2) \times (d \times a_3)
$$

where $a = a_1 \times a_2 \times a_3$.

For consecutive integers with $a=8$ (split into $2\times2\times2$):

$$
8 \times 9 \times 10 \times 11 = (9\times2) \times (10\times2) \times (11\times2) = 18 \times 20 \times 22 = 7920
$$

Original: $8 \times 9 \times 10 \times 11 = 7920$. ✓

For $a=27$ (split into $3\times3\times3$):

$$
27 \times 28 \times 29 \times 30 = (28\times3) \times (29\times3) \times (30\times3) = 84 \times 87 \times 90 = 657720
$$
