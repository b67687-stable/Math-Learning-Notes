## Layer 1: First Integer Patterns

Simple observations about integers that require only Layer 0.

### `discrete-mathematics/logic/consecutive-integer-multiples-are-divisible-by-2.ipynb`
- Consecutive integers differ by 1 (odd number) → one must be even, one odd
- Therefore $k(k+1)$ is always even; $k(k+1)(k+2)\dots$ always has an even factor

### `discrete-mathematics/number-theory/divisibility/number-composition.ipynb`
- Base-10 decomposition: $N = d_k \cdot 10^k + d_{k-1} \cdot 10^{k-1} + \dots + d_1 \cdot 10 + d_0$
- Each digit $0 \le d_i \le 9$

### `discrete-mathematics/number-theory/divisibility/sum-of-digits-is-less-than-or-equal-to-its-number.ipynb`
- For single-digit: number = digit sum.
- For multi-digit: every place except ones multiplies by $\ge 10$, so sum of digits $\le$ number.

### `discrete-mathematics/number-theory/divisibility/divisibility-nomenclature.ipynb`
- "Divisible by n" means **fully divisible** (integer quotient)

### `discrete-mathematics/number-theory/fun-facts-about-factors.ipynb`
- Prime factorization $a = 2^k \cdot 3^j \cdot 5^l \cdot \dots$
- Every factor is a combination with exponents 0 through max
- Number of factors = $(k+1)(j+1)(l+1)\dots$

## Layer 2: The Divisibility Universe

A systematic exploration of divisibility rules. **Every rule uses the same algebraic skeleton**:

1. Write number as base-10 decomposition
2. Factor each power of 10 relative to the divisor $d$
3. Separate into "multiple of $d$" + "remainder expression"
4. The remainder expression IS the rule

### `discrete-mathematics/number-theory/divisibility/factoring-the-place-units.ipynb`
The general method: factor $d$ from either the digit coefficients or the place values. Since digits are unknown, place values are the tractable target.

### Individual rules (all in `divisibility/`):

| File | Divisor | Core Remainder Test | Pattern Type |
|---|---|---|---|
| `divisibility-by-2` | 2 | Ones digit $\in \{0,2,4,6,8\}$ | Last N digits |
| `divisibility-by-3` | 3 | Digit sum $\equiv 0 \pmod{3}$ | Digit sum |
| `divisibility-by-4` | 4 | Last 2 digits $\equiv 0 \pmod{4}$ | Last N digits |
| `divisibility-by-5` | 5 | Ones digit $\in \{0,5\}$ | Last N digits |
| `divisibility-by-6` | 6 | Pass rules for BOTH 2 and 3 | Composite |
| `divisibility-by-7` | 7 | $3\cdot$(rest) + ones $\equiv 0$ | Chop-off |
| `divisibility-by-8` | 8 | Last 3 digits $\equiv 0 \pmod{8}$ | Last N digits |
| `divisibility-by-9` | 9 | Digit sum $\equiv 0 \pmod{9}$ | Digit sum |
| `divisibility-by-10` | 10 | Ones digit $=0$ | Last N digits |
| `divisibility-by-11` | 11 | Alternating digit sum $\equiv 0$ | Chop-off / alternating |
| `divisibility-by-12` | 12 | Pass rules for BOTH 3 and 4 | Composite |
| `divisibility-by-13` | 13 | $3\cdot$(rest) $-$ ones $\equiv 0$ | Chop-off |

### `divisibility-by-powers-of-2.ipynb` — Generalization
$10^n = 2^n \cdot 5^n$, so divisibility by $2^t$ depends on the **last $t$ digits**.

### Cross-cutting insight
Three rule **families** emerge:
- **Last N digits** (2, 4, 5, 8, 10, powers of 2): divisor factorizes $10^n$
- **Digit sum** (3, 9): $10^n \equiv 1 \pmod{d}$
- **Chop-off recursive** (7, 11, 13): $10 = d \pm r$, substitute recursively

## Layer 3: The Arithmetic Engine

This is the **most-used tool** in the repo. Almost everything above Layer 3 depends on it.

### `discrete-mathematics/series-and-sequences/arithmetic-sum/arithmetic-sum.ipynb`
- **Core insight**: Arithmetic sequences are symmetric about their average
- Average of first and last = average of ALL terms (due to symmetry)
- Therefore: $S_n = \frac{\text{first} + \text{last}}{2} \cdot n$
- Which simplifies to: $\displaystyle S_n = \frac{n}{2}[2a + (n-1)d]$
- Also covers: rebuilding sequences from both sides (odd-length vs even-length)
- Special case: $1 + 2 + \dots + n = \frac{n(n+1)}{2}$

### `discrete-mathematics/series-and-sequences/arithmetic-sum/arithmetic-sum-generalisation.ipynb`
- General form: $\displaystyle \sum_{k=1}^{n} (ck + b)$ — any arithmetic sequence as constant multiple + displacement
- Direct formula: $\displaystyle \frac{n(2b + c(n+1))}{2}$

## Layer 4: First Fruits of Arithmetic Sums

Direct applications: plug into the arithmetic sum formula and simplify.

### `discrete-mathematics/series-and-sequences/sum-of-odd-numbers-is-square.ipynb`
$\displaystyle \sum_{k=1}^{n} (2k-1) = n^2$
- Each odd number is $2k-1$, sum via arithmetic formula → $n^2$

### `discrete-mathematics/series-and-sequences/sum-of-even-numbers-is-consecutive-multiple.ipynb`
$\displaystyle \sum_{k=1}^{n} 2k = n(n+1)$
- Factor out 2, use sum of natural numbers

### `discrete-mathematics/series-and-sequences/triangular-numbers-and-their-sum.ipynb`
- $T_n = \frac{n(n+1)}{2}$ defined
- $\displaystyle \sum_{k=1}^{n} T_k = \frac{n(n+1)(n+2)}{6}$ — derived via sum of consecutive multiples

### `discrete-mathematics/series-and-sequences/sum-of-consecutive-multiples.ipynb`
$\displaystyle \sum_{k=1}^{n} k(k+1) = \frac{n(n+1)(n+2)}{3}$
- Split into $\sum k^2 + \sum k$, combine
- Foundation for triangular number sum

### `discrete-mathematics/series-and-sequences/sum-of-odd-and-even-squares.ipynb`
- $\sum (2k)^2 = \frac{2n(n+1)(2n+1)}{3}$ (factor 4 out of $\sum k^2$)
- $\sum (2k-1)^2 = \frac{n(2n-1)(2n+1)}{3}$ (total $\sum k^2$ minus $\sum (2k)^2$)

## Layer 5: The Symmetric Sum Lens

A different *way of seeing* sums: building from the center outward.

### `symmetric-sums/symmetric-sum-of-odd-numbers.ipynb`
- Odd-length symmetric sum of odds $= 2n(n-1)+1$
- Even-length symmetric sum of odds $= 2n^2$
- Even-length odds = odd-length evens (a duality)

### `symmetric-sums/symmetric-sum-of-even-numbers.ipynb`
- Odd-length symmetric sum of evens $= 2n^2$
- Even-length symmetric sum of evens $= 2n(n+1)$

### `symmetric-sums/symmetric-sum-of-squares.ipynb`
- Odd-length: $\frac{n(2n^2+1)}{3}$
- Even-length: $\frac{n(n+1)(2n+1)}{3}$

### `symmetric-sums/symmetric-sum-of-cubes.ipynb`
- Odd-length: $\frac{n^2(n^2+1)}{2}$
- Even-length: $\frac{n^2(n+1)^2}{2}$

### `symmetric-sums/sum-of-odd-numbers-in-symmetric-sum.ipynb`
- **Visual dot-stacking proof**: $\sum (2k-1) = 1+2+\dots+n+\dots+2+1 = n^2$
- Each new odd number adds one dot to each existing column + two new end columns

### `symmetric-sums/sum-of-even-numbers-in-symmetric-sum.ipynb`
- Visual dot-stacking: $\sum (2k) = 1+2+\dots+n+n+\dots+2+1 = n(n+1)$

## Layer 6: Higher Powers

Climbing from $\sum k$ to $\sum k^2$, $\sum k^3$, and $\sum k^m$.

### Sum of squares — 3 methods (`sum-of-squares/`):

| Notebook | Method | Insight |
|---|---|---|
| `sum-of-squares-with-arithmetic-sum` | Expansion + regrouping | $k^2 = \underbrace{k+k+\dots+k}_{k\text{ times}}$, then sum columns as arithmetic sums → self-referential equation → $\frac{n(n+1)(2n+1)}{6}$ |
| `sum-of-squares-with-symmetric-sum` | Symmetric decomposition | $k^2 = 1+2+\dots+k+\dots+2+1$, then sum as $T_n + T_{n-1}$ |
| `sum-of-squares-with-stack-of-symmetric-sums` | Vertical stacking | Stack symmetric sums, sum columns → arithmetic sums → solve |

### Sum of cubes (`sum-of-cubes/`):
| Notebook | Method |
|---|---|
| `sum-of-cubes-with-arithmetic-sum` | From the arithmetic sum expansion of each cube |
| `sum-of-cubes-with-symmetric-sums` | From symmetric sum decomposition |
| `sum-of-cubes-with-sum-of-squares` | From the known sum of squares formula |

### `cubes-as-sum-of-consecutive-odd-numbers.ipynb`
$n^3 =$ sum of $n$ consecutive odd numbers
- Takes difference of successive sums-of-cubes (which equal $(T_n)^2$)
- Strips away shared odd terms → isolates the $n$ new odd numbers

### `sums-of-powers.ipynb`
- **Recursive approach** (precursor to Faulhaber's formula)
- Expand $(k+1)^{n+1}$ via binomial theorem
- Telescope the sum → express $\sum k^{n}$ in terms of all lower $\sum k^{m}$
- This is how you'd derive $\sum k^4, \sum k^5, \dots$ systematically

## Layer 7: Pattern-Generating Mechanisms

These notebooks don't just *derive a formula* — they uncover a *mechanism* that generates infinite families of identities.

### `sharing-and-splitting.ipynb` + `ai-extensions/sharing-and-splitting/`
**Partition Regrouping**:
- Split a term into $n-1$ parts, distribute → sum preserved, term count drops by 1
- **Uniform arithmetic case**: $a = kd(n-1)$ generates nice patterns
- **Original pattern** ($k=m$): each row starts at $m^2$
- **Meta-pattern**: any sequence $f(m)$ generates a valid sharing identity

Pattern families discovered:
| $f(m)$ | Starting term | Who |
|---|---|---|
| $1$ | $m$ | Simplest (k-constant) |
| $m$ | $m^2$ | Original squares |
| $m^2$ | $m^3$ | Cubes |
| $2m-1$ | $m(2m-1)$ | Hexagonal numbers |
| $p_m$ | $m \cdot p_m$ | Primes |
| $F_m$ | $m \cdot F_m$ | Fibonacci |

### `telescoping-series.ipynb`
**Forward Difference Cancellation**:
- $\displaystyle \sum_{k=1}^{n} [f(k+d) - f(k)] = \sum_{\text{last }d} f(k) - \sum_{\text{first }d} f(k)$
- General theory: cancel intermediate terms, leaving only first $d$ and last $d$ terms
- Product version: $\displaystyle \prod \frac{f(k+d)}{f(k)}$ telescopes the same way
- **Unified view**: both additive and multiplicative telescoping emerge from the same structure

### `geometric-sum.ipynb`
**Self-Similarity**:
- $S_n = a + ar + ar^2 + \dots + ar^{n-1}$
- $S_n - rS_n = a(1-r^n)$ → $\displaystyle S_n = \frac{a(1-r^n)}{1-r}$
- For $|r| < 1$, $S_\infty = \frac{a}{1-r}$
- The self-similarity trick ($S - rS$) is itself a pattern-generating mechanism

## Layer 8: Derivative Series (Using Layer 7)

These apply the mechanisms from Layer 7 to specific series.

### `sum-of-reciprocal-consecutive-multiples.ipynb`
$\displaystyle \sum_{k=1}^{n} \frac{1}{k(k+1)} = \frac{n}{n+1}$
- Partial fractions: $\frac{1}{k(k+1)} = \frac{1}{k} - \frac{1}{k+1}$
- **Telescopes** → $1 - \frac{1}{n+1} = \frac{n}{n+1}$

### `sum-of-reciprocal-triangle-numbers.ipynb`
$\displaystyle \sum_{k=1}^{n} \frac{1}{T_k} = \frac{2n}{n+1}$
- Factor 2: $\frac{2}{k(k+1)} = 2(\frac{1}{k} - \frac{1}{k+1})$
- **Telescopes** the same way

Techniques useful everywhere.

### `algebra/quadratic-slide-and-divide.ipynb`
- Factoring $ax^2 + bx + c$ when $a \ne 1$
- **Trick**: multiply $c$ by $a$, factor $ax$ form, then divide roots by $a$
- Why it works: $ax^2 + bx + c = \frac{(ax)^2 + b(ax) + ac}{a}$

## What Your Repo Reveals About Your Thinking

Looking at the full map, a very consistent pattern emerges in **how** you explore math:

1. **Anchor in fundamentals** — Every new topic starts with definitions, not formulas
2. **Find the mechanism** — You don't stop at "what's the answer?", you ask "what's the process?" (sharing, telescoping, symmetric building, self-similarity)
3. **Multiple lenses** — Sum of squares has 3 derivations; cubes has 3 more. You collect perspectives until the idea is fully owned
4. **Generalize relentlessly** — Every specific result gets pushed to its most general form
5. **Connect across domains** — Divisibility rules, symmetric sums, arithmetic engine, sharing mechanism — they all talk to each other

The repo is not a textbook. It's a **personal mathematical universe** where every idea has its place in a web of dependencies, and nothing is accepted on authority.
