# The Composition of a Number

$$
\text{Number} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n places}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

Lets look into the `powers of 10`

# Factoring Powers of 10

Lets start from $10^0$
- Which equals `1`

$$
10^{0} = 1
$$

We cannot `partition` nor `factor` this that so lets move on to $10^1$

## Intentional Partitioning

Notice that `10` can be `partitioned` into `2 fives`

$$
10 = 5 \cdot 2
$$

If we move on to the `next power of 10`, we multiply by `10` to get this

$$
100 = 5 \cdot 20
$$

Noticing the `10` in the partition, we can split it like how we did before

$$
1000 = 5 \cdot 200
$$


Thus, we get this pattern that replicates itself

$$
\begin{align}
    10 &= 5 \cdot 2 \\
    100 &= 5 \cdot 20 \\
    1000 &= 5 \cdot 200 \\
    &\vdots \\
\end{align}
$$

## Generalisation of Such Partitioning

Using this pattern, where we `multiply by 10` and `group by 5`
- All the `non-negative powers of 10` can be partitioned like so

$$
\boxed{
    10^{n} = 5 \biggr( 2 \cdot 10^{n-1} \biggr)
}
$$

# Factoring the Number

This is the structure of a number

$$
\text{Number} = \underbrace{\overline{k \space m \space \cdots \space c \space b \space a}}_{\text{n places}} =
\space \underline{k \cdot 10^{n-1}} + 
\underline{m \cdot 10^{n-2}} + 
\cdots + 
\underline{c \cdot 10^{2}} + 
\underline{b \cdot 10^{1}} + 
\underline{a \cdot 10^{0}}
$$

Since we know this

$$
10^{n} = 5 \biggr( 2 \cdot 10^{n-1} \biggr)
$$

We factor it in

$$
\text{Number} = 
5 \biggr( 2 \cdot 10^{n-2} \biggr) k + 
5 \biggr( 2 \cdot 10^{n-3} \biggr) m + 
\cdots + 
5 \biggr( 2 \cdot 10^{1} \biggr) c + 
5 \biggr( 2 \cdot 10^{0} \biggr) b + 
a 
$$

Factor out the constant `5`, and `2` for making the equation simpler, and we get

$$
\boxed{
    \text{Number} = 
    5 \Biggr(
        2 \biggr(
            10^{n-2} k + 
            10^{n-3} m + 
            \cdots + 
            10c + 
            b 
        \biggr)
    \Biggr) + 
    a 
}
$$



# Conclusion

Thus, we see that `this part` of the number is a `multiple of 5`
- Thus making it `divisible by 5`

$$
\textcolor{lightgray}{\text{Number} =} 
5 \Biggr(
    2 \biggr(
        10^{n-2} k + 
        10^{n-3} m + 
        \cdots + 
        10c + 
        b 
    \biggr)
\Biggr) \textcolor{lightgray}{+ 
a}
$$

Thus for the `whole number to be divisible by 5`,  this part also has to be a `multiple by 5`

$$
\textcolor{lightgray}{\text{Number} = 
5 \Biggr(
    2 \biggr(
        10^{n-2} k + 
        10^{n-3} m + 
        \cdots + 
        10c + 
        b 
    \biggr)
\Biggr)} + 
a
$$

Realise that it is the `digit in the ones place`

$$
\textcolor{lightgray}{
    \text{Number} \space = \space} 
\overline{\textcolor{lightgray}{k \space m \space \cdots \space c \space b} \space a}
$$

Thus, to determine if the whole number is `divisible by 5`,  the digit in the `ones place` has to be a `multiple of 5`


There are only `2 digits` that is a `multiple of 5`
- `0` itself, as it is a multiple of all numbers, and the `0-th multiple of 5`
- and  `5`, the `first multiple of 5`

All the other multiples of 5 are `too big` to be a `positive single digit`

$$
\text{All digits that are a multiple of 5:} \quad 0, \space 5
$$

# Divisibility by 5 Rule

Thus, to determine if a number is `divisible by 5`

$$
\boxed{
    \text{ The digit in the ones place has to be 0 or 5}
}
$$
